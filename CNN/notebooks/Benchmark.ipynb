# Benchmark 

```python
import onnx
import onnxruntime as ort
import numpy as np
from PIL import Image
import glob
import sys
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Ce notebook prends différents "arguments" que vous devez remplir avant de lancer le benchmark 
# Ces arguments sont les suivants
# A directory witch contains model onnx
# A directory which contains images
```

```python
model_dir = "../../CNN/model/500*500/*"
images_dir = "../../../datasets/original_500_500/test/*"
```

```python
result = []
models = []
list_model_path = glob.glob(model_dir)

list_dir_path = glob.glob(images_dir)
list_image_path = []
for index in range(len(list_dir_path)):
    list_image_path.append(glob.glob(list_dir_path[index] + "/*"))
    list_dir_path[index] = list_dir_path[index].split("/")[-1]
```

```python
for model in list_model_path:
    n_total = 0
    n_succed = 0
    ort_sess = ort.InferenceSession(model)
    for index in range(len(list_dir_path)):
        for image_path in list_image_path[index]:
            img = np.array(Image.open(image_path),dtype=np.float32)
            img = img.reshape(1, 500, 500, 1)
            name = image_path.split("/")
            n_total += 1
            img = np.array(Image.open(image_path),dtype=np.float32)
            img = img.reshape(1, 500, 500, 1)
            outputs = ort_sess.run(None, {'conv2d_input': img})
            if float(outputs[0]) >= 0.5 and list_dir_path[index][0:4] == "PNEU":
                n_succed += 1
            elif float(outputs[0]) <= 0.5 and list_dir_path[index][0:4] == "NORM":
                n_succed += 1
    result.append(n_succed)
    models.append(model.split("/")[-1])
    print("Model : ", model)
    print("Success : ", str(n_succed), " on ", str(n_total), " évaluations")
    print("Accuracy : " + str(n_succed/n_total*100) + "%\n")
```

```python
df = pd.DataFrame({
    "models": models,
    "nb succeed": result
})
```

```python
#plt.bar(result["../../CNN/model/500*500/model-v1.onnx"], 624,20)
plt.figure(figsize=(25,12))
splot=sns.barplot(x="models", y="nb succeed", data=df)
plt.xlabel("models", size=16)
plt.ylabel("nb succeed", size=16)
plt.bar_label(splot.containers[0], label_type='center')
plt.show()
```
